```python
# -*- coding: utf-8 -*-
"""
Created on Wed Jun  2 19:26:49 2021

@author: mrasoulzadeh
"""
#these are programs/features in python that you have to import to use. Some may not become necessary/useful until later
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
import statsmodels.tsa.seasonal as sms
import warnings
import seaborn as sns
import scipy.stats
import time
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
###############################################################################
def readDataFramefromcsvfile(filename):
    df=pd.DataFrame(pd.read_csv(filename))
    df.columns = ["DATE", "TIME","WATER LEVEL"]
    #df=df.dropna(how='all')
    df=df.replace({'---': np.nan})
    df['WATER LEVEL']=df['WATER LEVEL'].astype(float)
   
    # indexNames = df[ df["WATER LEVEL"] == "---" ].index
    # # Delete these row indexes from dataFrame
    # df.drop(indexNames , inplace=True)
    
    df['DATE']=df['DATE'].astype('datetime64[ns]')
    df=df.groupby('DATE').mean()  # average for waterlevel data

    df.reset_index()
    return df
    
#print(pd.DataFrame(pd.read_csv("Baldwin2.csv")))
```

```python
#------------------------------------------------------------------------------
def mergeWL_PPT (df_WL,df_PPT,stationname):
    ppt=df_PPT[df_PPT["STATION"].isin([stationname])]
    df=pd.merge(df_WL,ppt, on='DATE')
    #df.set_index("DATE", inplace=True)

    # add missing dates
    r = pd.date_range(start=df["DATE"].min(), end=df["DATE"].max())
    df=df.set_index('DATE').reindex(r).fillna('---').rename_axis('DATE').reset_index()    
#df[SumPRCP] = df

#print(ppt[["DATE"]])
#ppt['DATE']=ppt['DATE'].astype('datetime64[ns]')
#ppt['YEAR'] = ppt["DATE"].dt.year
#print(ppt[["YEAR"]])
#ppt2=ppt.groupby('YEAR').sum()
#ppt2=ppt2[["PRCP"]]

    df.loc[df['YEAR'] == 2010 , 'SumPRCP'] = '17.31' 
    df.loc[df['YEAR'] == 2011 , 'SumPRCP'] = '416.66' 
    df.loc[df['YEAR'] == 2012 , 'SumPRCP'] = '373.43' 
    df.loc[df['YEAR'] == 2013 , 'SumPRCP'] = '495.90' 
    df.loc[df['YEAR'] == 2014 , 'SumPRCP'] = '429.05' 
    df.loc[df['YEAR'] == 2015 , 'SumPRCP'] = '475.39' 
    df.loc[df['YEAR'] == 2016 , 'SumPRCP'] = '355.39' 
    df.loc[df['YEAR'] == 2017 , 'SumPRCP'] = '480.00' 
    df.loc[df['YEAR'] == 2018 , 'SumPRCP'] = '497.58' 
    df.loc[df['YEAR'] == 2019 , 'SumPRCP'] = '421.28' 
    df.loc[df['YEAR'] == 2020 , 'SumPRCP'] = '240.44' 

    df['DAY'] = df["DATE"].dt.day
    df['MONTH'] = df["DATE"].dt.month
    df['YEAR'] = df["DATE"].dt.year
    df['WATER_YEAR'] = df["DATE"].dt.year.where(df["DATE"].dt.month < 10, df["DATE"].dt.year + 1)
    df['SEASON'] = df["DATE"].dt.month%12 // 3 + 1
    df=df[~((df['MONTH']==2)&(df['DAY']==29))] # drop leap day
    df["STATION"]=stationname
    if ~(df["WATER LEVEL"][df["WATER LEVEL"]=='---'].count()==0):
        df["WATER LEVEL"]=df["WATER LEVEL"].replace({'---': np.nan})
    df["WATER LEVEL"]=df["WATER LEVEL"].interpolate(method='cubic')
    df["WATER LEVEL"]=df["WATER LEVEL"].interpolate(method='linear',limit_direction='backward' , limit=4 )
    
    if ~(df["PRCP"][df["PRCP"]=='---'].count()==0):
        df["PRCP"]=df["PRCP"].replace({'---': 0})   
    return df
```

```python
#------------------------------------------------------------------------------
def plot_WL_PCPT (df,stationname):
    fig = plt.figure(num=None, figsize=(15, 8), dpi=80)
    ax = fig.add_subplot(1, 1, 1)
    plt.grid(b=True, which='major', color='black', linewidth='0.5', linestyle='-',axis='x')
    plt.xlabel('Time [yr]', fontsize=20)
    plt.ylabel('Water table depth (ft)', fontsize=20)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.ylim(df['WATER LEVEL'].min()-1,  df['WATER LEVEL'].max()+10)
    ax_2 = ax.twinx()
    ax.plot(df['DATE'], df['WATER LEVEL'], label = "Water Level", linewidth=3, color='r')
    ax_2.plot(df['DATE'], df['PRCP'], label = "PRCP", linewidth=2)
    plt.ylim(0, df['PRCP'].max())
    plt.gca().invert_yaxis()
    lines, labels = ax.get_legend_handles_labels()
    lines_2, labels_2 = ax_2.get_legend_handles_labels()
    lines = lines + lines_2
    labels = labels + labels_2
    ax.legend(lines, labels, loc='lower right', fontsize=15)
    plt.ylabel('Precipitation (in)', fontsize=20)
    plt.title(stationname, fontsize=20)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=15)
    plt.grid()
    plt.show()
    return
```

```python
#------------------------------------------------------------------------------
def getOneColumn (df,datafieldName):
    mg=df.groupby("WATER_YEAR")
    df_WLMergedByMonthDay=pd.DataFrame()
    for mjd in mg.groups.keys():
        dfYear=mg.get_group(mjd)
        if (len(dfYear)==365):
            if df_WLMergedByMonthDay.empty:
                df_WLMergedByMonthDay=dfYear
                df_WLMergedByMonthDay=df_WLMergedByMonthDay[['DAY','MONTH','DATE',datafieldName]]
            else:
                df_WLMergedByMonthDay = df_WLMergedByMonthDay.append(dfYear[['DAY','MONTH','DATE',datafieldName]])
    return df_WLMergedByMonthDay
```

```python
#------------------------------------------------------------------------------
def getFieldGroups (df,groupfieldName,datafieldName):
    mg=df.groupby(groupfieldName)
    df_WLMergedByMonthDay=pd.DataFrame()
    for mjd in mg.groups.keys():
        dfYear=mg.get_group(mjd)
        if df_WLMergedByMonthDay.empty:
            df_WLMergedByMonthDay=dfYear
            df_WLMergedByMonthDay=df_WLMergedByMonthDay[['DAY','MONTH',datafieldName]]
            df_WLMergedByMonthDay.columns=['DAY','MONTH',datafieldName + str(mjd)]

        else:
            df_WLMergedByMonthDay = df_WLMergedByMonthDay.merge(dfYear[['DAY','MONTH',datafieldName]], how='outer', left_on=['DAY','MONTH'], right_on=['DAY','MONTH'])
            df_WLMergedByMonthDay = df_WLMergedByMonthDay.rename({datafieldName: datafieldName + str(mjd)}, axis=1)
        #df_WLMergedByMonthDay=dfMergedByMonthDay.merge(dfYear["WATER LEVEL"], how='inner', on=['MONTH', 'DAY']))
    return df_WLMergedByMonthDay
```

```python
#------------------------------------------------------------------------------
def getOneColumnData (df):
    mjd=pd.DataFrame()
    for i in range(3,len(df.columns)-1):
        mjd2=pd.DataFrame(df.iloc[:,i])
        mjd2.columns = ["VALUE"]
        if mjd.empty:
            mjd=mjd2;
        else:
            mjd=mjd.append(mjd2,ignore_index=True)
    mjd=mjd.replace(np.nan, 0)
    return mjd
```

```python
#------------------------------------------------------------------------------
def annot_max(x,y, ax=None):
    ymax = y.max()
    xmax = x[np.argmax(y)]
        
    text= "x={:.3f}, y={:.3f}".format(xmax, ymax)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax.annotate(text, xy=(xmax, ymax), xytext=(0.94,0.96), **kw)
    return
```

```python
#------------------------------------------------------------------------------
def mround(x, m=5):
    '''Helper method for multiple round'''
    return int(m * round(float(x)/m))
```

```python
#------------------------------------------------------------------------------
def heatmap(corr_matrix):
    ax = sns.heatmap(
        corr_matrix, 
        vmin=-1, vmax=1, center=0,
        cmap=sns.diverging_palette(20, 220, n=200),
        square=True)
    return
```

```python
#------------------------------------------------------------------------------
# Time lagged cross correlation
def crosscorr(datax, datay, lag=0):
    """ Lag-N cross correlation. 
    Shifted data filled with NaNs 

    Parameters
    ----------
    lag : int, default 0
    datax, datay : pandas.Series objects of equal length
    Returns
    ----------
    crosscorr : float
    """ 
    return datax.corr(datay.shift(lag))   
```
